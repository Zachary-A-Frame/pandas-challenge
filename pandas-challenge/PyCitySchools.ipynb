```python
merged_df = students_df.merge(schools_df, how = 'left', on = 'school_name')

merged_df
```

```python
grouped_schools = merged_df.set_index('school_name').groupby(['school_name'])

school_types = schools_df.set_index('school_name')['type']

total_students_by_school = grouped_schools['Student ID'].count()

school_budget = schools_df.set_index('school_name')['budget']

student_budget = schools_df.set_index('school_name')['budget'] / schools_df.set_index('school_name')['size']

average_math = grouped_schools['math_score'].mean()/100
average_reading = grouped_schools['reading_score'].mean()/100

passing_math_scores = merged_df[merged_df['math_score'] >= 70].groupby('school_name')['Student ID'].count()/total_students_by_school
passing_reading_scores = merged_df[merged_df['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/total_students_by_school
overall = merged_df[(merged_df['reading_score'] >= 70) & (merged_df['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/total_students_by_school
```

```python
school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": total_students_by_school,
    "Per Student Budget": student_budget,
    "Total School Budget": school_budget,
    "Average Math Score": average_math,
    "Average Reading Score": average_reading,
    '% Passing Math': passing_math_scores,
    '% Passing Reading': passing_reading_scores,
    "Overall Passing Rate": overall
})

```

```python
school_summary.style.format({
    "Total Students":"{:,}",
    "Per Student Budget":"${:.0f}",
    "Total School Budget": "${:,}",
    "Average Math Score": "{:.1%}",
    "Average Reading Score": "{:.1%}",
    "% Passing Math": "{:.1%}",
    "% Passing Reading": "{:.1%}",
    "Overall Passing Rate": "{:.1%}"
})
```

```python
# sort values by passing rate and then only print top 5

top_performers = school_summary.sort_values("Overall Passing Rate", ascending = False)
top_performers.head().style.format({'Total Students': '{:,}',
                           "Total School Budget": "${:,}", 
                           "Per Student Budget": "${:.0f}", 
                           "% Passing Math": "{:.1%}", 
                           "% Passing Reading": "{:.1%}", 
                           "Overall Passing Rate": "{:.1%}"})
```
